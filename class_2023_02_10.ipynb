```python
df = pd.read_csv("https://sta663-sp23.github.io/slides/data/us_rent.csv")
```

```python
( df
  .pivot(
    index="name", columns="variable", values="estimate"
  )
  .reset_index()
  .assign(
    ratio = lambda d: np.round(d.income / d["rent"],2)
  )
  .sort_values("ratio", ascending=True)
)
```
